```python
# Pyplot importieren
import matplotlib.pyplot as plt

# Seaborn importieren
import seaborn as sns
import pandas as pd

# Farben von Pyplot importieren
from matplotlib.colors import ListedColormap

# Numpy importieren
import numpy as np

# Datasets, Klassifikatoren und Beurteilungshelfer (Metriken importieren)
from sklearn import tree, metrics, neighbors, datasets
from sklearn.model_selection import train_test_split
```

Datenset laden

```python
df_mpg = sns.load_dataset('mpg')
df_mpg.head(10)
```

```python
df_mpg.corr()
```

Datenset in X und Y aufteilen, Test-Train-Split erstellen

```python
X = df_mpg[['mpg', 'acceleration']]
Y = df_mpg['cylinders']

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, shuffle=True
)
```

Modell erstellen und trainieren

```python
clf = neighbors.KNeighborsClassifier(15, weights='uniform')
clf.fit(X_train, y_train)
```

Schätzungen erstellen

```python
y_predict = clf.predict(X_test)
```

Vergleichsliste von Schätzungen und Testwerten ausgeben

(Dieser Schritt wird durch die Evaluierung später ersetzt)

```python
pd.DataFrame([y_predict, y_test]).transpose().head(20)
```

Zweites Modell erstellen und trainieren

```python
clf = tree.DecisionTreeClassifier(max_depth=20)
clf.fit(X_train, y_train)
```

Zweite Schäztungen erstellen

```python
y_predict = clf.predict(X_test)
```
